# Implementation with Numpy

```python
import numpy as np
```

```python
# N is batch size; D_in is input dimension;
# H is hidden dimension; D_out is output dimension.
N, D_in, H, D_out = 64, 1000, 100, 10
```

```python
# Create random input and output data
x = np.random.randn(N, D_in)
y = np.random.randn(N, D_out)
```

```python
# Randomly initialize weights
w1 = np.random.randn(D_in, H)
w2 = np.random.randn(H, D_out)
```

```python
learning_rate = 1e-6
```

```python
for epoch in range(1000):
    # forward pass : compute y_hat
    h = np.dot(x, w1)
    h_relu = np.maximum(h, 0)
    y_pred = np.dot(h_relu, w2)
    
    # Compute loss:
    loss = np.sum(np.square(y_pred-y))
    print(epoch, loss)
    
    # Backpropogations
    #grad_y_pred = 2*np.sum(y-y_pred)
    #grad_w2 = 2*np.sum(y-y_pred)*h_relu
    #grad_h_relu = 2*np.sum(y-y_pred)*w2
    #grad_h = 2*np.sum(y-y_pred)*w2
    #grad_w1 = 2*np.sum(y-y_pred)*w2*x
    grad_y_pred = 2.0 * (y_pred - y)
    grad_w2 = h_relu.T.dot(grad_y_pred)
    grad_h_relu = grad_y_pred.dot(w2.T)
    grad_h = grad_h_relu.copy()
    grad_h[h < 0] = 0
    grad_w1 = x.T.dot(grad_h)
    
    #Update weights
    w1-=learning_rate*grad_w1
    w2-=learning_rate*grad_w2
    
```

# Implementation with PyTorch

```python
import torch
```

```python
dtype = torch.float
device = torch.device("cpu")
```

```python
N = 64
D_in = 1000
H = 100
D_out = 10
```

```python
x = torch.randn(N, D_in, device=device, dtype=dtype)
y = torch.randn(N, D_out, device=device, dtype = dtype)
```

```python
w1 = torch.randn(D_in, H, device=device, dtype=dtype, requires_grad=True)
w2 = torch.randn(H, D_out, device=device, dtype=dtype, requires_grad=True)
```

```python
for epoch in range(5000):
    y_pred = x.mm(w1).clamp(min=0).mm(w2)
    loss = (y_pred-y).pow(2).sum()
    if epoch % 1000 == 0:
        print(epoch, loss.item())
    
    loss.backward()
    
    with torch.no_grad():
        w1 -= learning_rate * w1.grad
        w2 -= learning_rate * w2.grad
        
        w1.grad.zero_()
        w2.grad.zero_()
```

# Using nn module - Functional 

```python
from torch.nn import Linear
from torch.nn.functional import relu
```

```python
class LinearRegression(torch.nn.Module):
    def __init__(self, D_in, H, D_out):
        super(LinearRegression, self).__init__()
        self.h1 = Linear(in_features =D_in,
                         out_features = H)
        self.h2 = Linear(in_features = H,
                             out_features = D_out)
    def forward(self, x):
        h_1 = self.h1(x)
        h_1_relu = relu(h_1)
        y_pred = self.h2(h_1_relu)
        return y_pred
```

```python
x = torch.from_numpy(np.random.randn(64, 1000))
y = torch.from_numpy(np.random.randn(64, 10))
```

```python
model = LinearRegression(D_in, H, D_out)
```

```python
criterion = torch.nn.MSELoss(reduction="sum")
optimizer = torch.optim.SGD(model.parameters(), lr = 1e-4)
```

```python
for epoch in range(1000):
    y_pred = model(x.float())
    loss = criterion(y_pred, y.float())
    if epoch%100 ==0 :
        print(epoch, loss.item())
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
```

# Using nn module - Sequential API

```python
from torch import nn
from torch import functional as F
```

```python
model = nn.Sequential(
            nn.Linear(D_in, H),
            nn.ReLU(),
            nn.Linear(H, D_out)
            )
```

```python
criterion = nn.MSELoss(reduce="sum")
optimizer = torch.optim.SGD(params = model.parameters(), lr=1e-4)
```

```python
for epoch in range(1000):
    y_pred = model(x.float())
    loss = criterion(y_pred, y.float())
    optimizer.zero_grad()
    
    loss.backward()
    optimizer.step()
    if epoch%100 == 0:
        print(epoch, loss.item())
```
